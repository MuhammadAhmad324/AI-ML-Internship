```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Better plot style
sns.set_style("whitegrid")
```

```python
# Load Iris Dataset

df = sns.load_dataset('iris')

# Display first 5 rows
df.head()
```

```python
print("Dataset Shape:")
print(df.shape)
```

```python
print("Column Names:")
print(df.columns)
```

```python
print("First 5 Rows:")
print(df.head())
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x='sepal_length',
    y='petal_length',
    hue='species'
)

plt.title("Sepal Length vs Petal Length")
plt.show()
```

```python
df.hist(figsize=(10,8))
plt.suptitle("Feature Distributions")
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.boxplot(data=df)

plt.title("Box Plot of Iris Features")
plt.show()
```

```python
sns.pairplot(df, hue='species')
plt.show()
```

Insights:

1. Iris dataset contains 150 samples and 5 columns.
2. No missing values were found.
3. Petal length and petal width show strong correlation.
4. Setosa species is clearly separable from other species.
5. Few potential outliers can be observed in sepal width.

Conclusion

The Iris dataset was successfully loaded and explored using Pandas.
Various visualizations such as scatter plots, histograms, and box plots were created using Matplotlib and Seaborn.

The analysis showed clear differences among Iris species, especially in petal measurements. No missing values were found, making the dataset clean and suitable for machine learning tasks.
